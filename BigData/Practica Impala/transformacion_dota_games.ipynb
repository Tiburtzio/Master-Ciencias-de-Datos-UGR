```python
import pandas as pd
games_modes = ['All Pick','Captain\'s Mode','Random Draft','Single Draft',
              'All Random', 'Intro', 'Diretide',
              'Reverse Captain\'s Mode ', 'The Greeviling']
games_type = ['Private Game','Unranked Game',"Ranked Game"]
teams = ['Red Team',"Blue Team"]
player = ['Red Team',"None","Blue Team"]

columns = ['WinnerTeam',"ClusterID","GameMode","GameType"]
columns.extend(hero_names)

df = pd.read_csv('dota2Test.csv',header=None)
df.columns = columns
df.head(10)
np.unique(df['antimage'])
```

```python
def transfromToCategorical(series,categories):
    unique_values = np.unique(series)
    for ind in range(len(unique_values)):
        unique_val = unique_values[ind]
        series[series == unique_val] = categories[ind]
    
    return series
```

```python
nuevo_winner = transfromToCategorical(df["WinnerTeam"].copy(),
                                     teams)
nuevo_modes = transfromToCategorical(df["GameMode"].copy(),
                                    games_modes)
nuevo_types = transfromToCategorical(df["GameType"].copy(),
                                    games_type)

df["WinnerTeam"] = nuevo_winner
df["GameMode"] = nuevo_modes
df["GameType"] = nuevo_types

df.head(10)
```

```python
aux = transfromToCategorical(df["axe"].copy(),
                            player)
```

```python
heroes = df.columns.values[4:]
for hero in heroes:
    aux = transfromToCategorical(df[hero].copy(),
                                player)
    df[hero] = aux
    aux = []
    
df.head(10)
```
